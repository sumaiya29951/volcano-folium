```python
import folium
import pandas
```

```python
data= pandas.read_csv("Volcano.txt")

latitude=list(data["LAT"])
longitude=list(data["LON"])
elevation_data=list(data["ELEV"])
```

```python
map = folium.Map(location=[38, -100], zoom_start=6, tiles="Stamen Terrain")

feature_group = folium.FeatureGroup(name="My Map")
```

```python
def color_icon(elv_data):
    if elv_data < 1000:
        return 'green'
    
    elif 1000 <= elv_data < 3000:
        return 'blue'
    
    else:
        return 'red'
    

for lat, lon, elv in zip(latitude, longitude, elevation_data):
    feature_group.add_child(folium.Marker(
        location =[lat, lon], popup =str(elv)+"m", icon =folium.Icon(color=color_icon(elv))))
    
map.add_child(feature_group)

map
```
